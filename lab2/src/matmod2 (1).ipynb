# 1. Объект исследования.
Объект исследования - система "маятник-пуля". Практическая задача, которую мы решаем - определение скорости пули, по отклонению маятника. В рамках исследуемого процесса, к маятнику прикреплено орудие, которое совершает выстрел в горизонтальном направлении.

# 2. Содержательная постановка задачи.

Наша модель должна давать количественный ответ - скорость, с которой была выпущена пуля из орудия.

Не имея под рукой специальной лаборатории, воспользуемся для определением скорости пули относительно простым устройством - баллистическим маятником.

Для построения модели воспользуемся двумя фундаментальными законами природы - законом сохранения энергии и законом сокранения импульса.

Входные параметры:

$M$ - совокупная масса орудия и маятника, $m$ - масса пули.

$l, L$ - длина нити маятника, и максимальная величина смещения маятника по горизонтали.

$M$, $m$ и $l$ известны до запуска пули, и определяются из характеристик установки. $L$ определяется в результате выстрела. Однако в рамках модели эти параметры удобно учитывать однородно.

Выходной параметр - начальная скорость пули $u$.

# 3. Концептуальная постановка задачи.
image.png

Решать задачу мы будем в рамках классической механики.

Систему будем считать замкнутой в горизонтальном направлении, при выполнении следующих условий:

1) Вектор скорости пули должен быть направлен в направлении оси x, по прямой, проходящей через центр тяжести маятника с орудием. Последние условие может выполниться лишь приближенно, для этого характерный размер маятника должен быть на порядок меньше длины нити.

2) Время выстрела должно быть заметно меньше периода собственных колебаний маятника.

Эти условия на практике могут быть обеспечены выбором достаточно длинной нити подвеса маятника.

Баллистический маятник - прибор, применяемый для измерения начальных скоростей пули или снаряда. В данном случае он представляет из себя тяжелый металлический цилиндр, заполненный вязким веществом и подвешенный на четырех нерастяжимых нитях. К цилиндру сбоку прикреплено простое пружинное орудие. В результате выстрела по закону отдачи маятник с орудием приобретают некоторую начальную скорость, и затем отклоняются на некоторое расстояние. По этому расстоянию можно определить скорость пули.



# 4. Математическая постановка задачи.

## Закон сохранения энергии.
image.png

Пуля массой m, вылетая со скоростью $v$, сообщит маятнику с орудием по закону отдачи кинетическую энергию $\frac {u^2m} {2}$. В результате отклонения маятник с орудием поднимется на высоту $h$, определяемую из закона сохранения механической энергии $Mgh=\frac {u^2m} {2}$. Отсюда:
$$u = \sqrt{\frac {2M}{m}gh}$$

Определим зависимость высоты $h$ от отклонения L. Ее можно выразить через угол отклонения $\alpha$:

$$h = l(1 - \cos\alpha) = 2l\sin^2\frac{\alpha}{2}$$

Виду того, что длина нитей l на порядок превосходит смещение маятника L, и угол отклонения маятника достаточно мал, то можно воспользоваться эквивалентностью:

$$\sin\frac{\alpha}{2}\approx \frac{\alpha}{2}
\approx \frac {tg\alpha}{2}=\frac {L}{2l}$$

Таким образом, высота $h$ приближенно равна:

$$h\approx\frac {L^2}{2l}$$

Итоговый вид зависимости скорости пули от отклонения примет вид:

$$u=L\sqrt{\frac{Mg}{ml}} $$



## Закон сохранения энергии и закон сохранения импульса.

image.png

В начальный момент времени маятник с орудием приобретают скорость $V$. Она определяется из уравнения закона сохранения энергии:

$$M\frac{V^2}{2}=Mgh \\
 V = \sqrt{2gh}\approx L\sqrt{\frac{g}{l}}$$

 Скорость полета пули находится из закона сохранения импульса, при условии, что на систему не действуют внешние силы:

 $$mu=MV \\
 u \approx \frac {M}{m}L\sqrt{\frac {g}{l}}$$

# 5. Качественный анализ и проверка корректности модели.

С математической точки зрания задача свелась к решению пары линейных уравнений, по одному для каждого метода.

Из закона сохранения энергии:

$$u=\sqrt{\frac{M}{m}}L\sqrt{\frac{g}{l}} $$

Из законов сохранения импульса и энергии:

$$u = \frac {M}{m}L\sqrt{\frac {g}{l}}$$



Выполним контроль размерности уравнений:

$$u=\sqrt{\frac{M}{m}}L\sqrt{\frac{g}{l}} \;\;\;\;\;⇒\;\;\;\; \left[\frac{м}{с}\right] = \sqrt{\left[\frac{кг}{кг}\right]} \left[м\right] \sqrt{\left[\frac{м/с^2}{м}\right]} \;\;\;\;\;⇒\;\;\;\; \left[\frac{м}{с}\right] = \left[\frac{м}{с}\right]$$

$$u=\frac{M}{m}L\sqrt{\frac{g}{l}} \;\;\;\;\;⇒\;\;\;\; \left[\frac{м}{с}\right] = \left[\frac{кг}{кг}\right] \left[м\right] \sqrt{\left[\frac{м/с^2}{м}\right]} \;\;\;\;\;⇒\;\;\;\; \left[\frac{м}{с}\right] = \left[\frac{м}{с}\right]$$

## Угол отклонения маятника ЗСЭ

В результате отклонения маятник с орудием поднимется на высоту $h$, определяемую из закона сохранения механической энергии

$$Mgh=\frac {u^2m} {2}$$
$$h = \frac{mu^2}{2Mg} $$

Определим зависимость угла отклонения $\alpha$ от высоты $h$ и длины нити $l$:

$$h = l(1 - \cos\alpha) = 2l\sin^2\frac{\alpha}{2}$$

Тогда
$$\alpha = 2\arcsin \left (\frac{u\sqrt{m}}{2\sqrt{Mlg}}\right )$$

$$h = l(1 - \cos\alpha) = 2l\sin^2\frac{\alpha}{2}$$

$$\alpha = 2\arcsin \left (\frac{V}{2\sqrt{lg}}\right )$$

$$mu=MV$$
Тогда
$$\alpha = 2\arcsin\left(\frac{um}{2M\sqrt{lg}}\right)$$

### 4) Поиск отклонения маятника из ЗСЭ и ЗСИ
Входные данные:
1. Масса маятника с пистолетом (M) [кг]
2. Масса пули (m) [кг]
3. Длина нити (l) [м]
4. Скорость пули (u) [м/с]

Использованные табличные значения:
1. Ускорение свободного падения (g) [9.8 м/с^2]

Выходные данные:
1. Угол отклонения маятника ($\alpha$) [градус]

Ограничения данных:
Массы $m$ и $M > 0$ 

Длина нити $l > 0$

Скорость пули $u >= 0$

Отклонение маятника $0 < L << l$  


```python
import numpy as np

g = 9.8

def velocity1(M, m, l, L):
    return L * np.sqrt(M * g / l / m)

def velocity2(M, m, l, L):
    return L * M / m * np.sqrt(g / l)
    
def angle1(M, m, l, u):
    return 2 * np.arcsin(u / (2 * np.sqrt(l * g)) * np.sqrt(m / M))
    
def angle2(M, m, l, u):
    return 2 * np.arcsin(u / (2 * np.sqrt(l * g)) * (m / M))
```

# 8. Проверка адекватности модели

Уточним параметры. Средняя масса пули - 0.01 кг. Масса матника - порядка 1 кг. Отклонение - порядка 0.05 метров. Длина нити - 1 м. Пружинный пистолет запускает пулю со скоростью порядка 10 - 15 м/с. Попробуем запустить программу, и посмотреть, что у нас выйдет.

```python
M = 1
m = 0.01

l = 2
L = 0.05

print(f"Скорость пули с учетом только закона сохранения энергии: {velocity1(M, m, l, L)} м/с.")
print(f"Скорость пули с учетом закона сохранения энергии и импульса: {velocity2(M, m, l, L)} м/с.")
```

Как видно, с учетом экспериментальных данных, модель, не учитывающая закон сохранения импульса, показывает результат значительно хуже. В то же время, модель, учитывающая оба закона, показывает вполне реалистичный результат.

# 9. Практическое использование построенной модели

Построим график зависимостей скорости пули от отклонения маятника, и график зависимости угла отклонения от скорости пули. Отклонение маятника будем брать в пределах от 2 до 10 см, скорость пули - от 10 до 20 м/c.

```python
import matplotlib.pyplot as plt

args = np.linspace(0.02, 0.1)

func = np.vectorize(lambda x: velocity1(M, m, l, x))

vals = func(args)
plt.figure(figsize=[7, 7])

plt.title("Зависимость скорости от отклонения без ЗСИ")
plt.xlabel("Отклонение, м")
plt.ylabel("Скорость пули, м/с")

plt.plot(args, vals);
plt.grid()
```

```python
args = np.linspace(0.02, 0.1)

func = np.vectorize(lambda x: velocity2(M, m, l, x))

vals = func(args)
plt.figure(figsize=[7, 7])

plt.title("Зависимость скорости от отклонения с учетом ЗСИ")
plt.xlabel("Отклонение, м")
plt.ylabel("Скорость пули, м/с")

plt.plot(args, vals);
plt.grid()
```

```python
args = np.linspace(10,20, 100)

func = np.vectorize(lambda x: angle1(M, m, l, x))

vals = np.rad2deg(func(args))
plt.figure(figsize=[7, 7])

plt.title("Зависимость угла отклонения от скорости без ЗСИ")
plt.xlabel("Скорость пули, м/с")
plt.ylabel("Отклонение, в градусах")

plt.plot(args, vals);
plt.grid()
```

```python
args = np.linspace(10,20, 100)

func = np.vectorize(lambda x: angle2(M, m, l, x))

vals = np.rad2deg(func(args))
plt.figure(figsize=[7, 7])

plt.title("Зависимость угла отклонения от скорости с учетом ЗСИ")
plt.xlabel("Скорость пули, м/с")
plt.ylabel("Отклонение, в градусах")

plt.plot(args, vals);
plt.grid()
```

Как видно, в случае, в котором не учитывается закон сохранения импульса, значение угла сильно завышена. Эта модель скорее показала свою неадекватность. В это же время, модель, учитывающая закон сохранения импульса, показывает реалистичные результаты.

## Дополнительный вопрос. 
### При каких значениях массы маятника выполняются предположения, что 

$$\sin \alpha \approx \alpha \\
\alpha \approx tg \alpha$$ 

Выполняется ли примерное равенство или нет - вопрос, который зависит от требуемой точности. Для того, чтобы требуемую точность можно было определять постфактум, введем дополнительный параметр максимальной относительной погрешности $p$. Допустим, для решаемой задачи, приемлимая точность - $p=0.05$.

Выполнение приближенного равенства будем проверять через следующее неравенство:

$$\dfrac {tg\alpha - \sin \alpha} {\sin \alpha} \leq p \\
\frac {1} {\cos \alpha} \leq p + 1 \\
\cos \alpha \geq \frac {1} {p + 1} \\
\alpha \leq \arccos \left( \frac {1} {p + 1} \right)$$

С другой стороны, в случае с формулой, учитывающей закон сохранения инерции, которая показала свою адекватность:

$$\alpha = 2\arcsin\left(\frac{um}{2M\sqrt{lg}}\right)$$

Сопоставляя, получаем:

$$2\arcsin\left(\frac{um}{2M\sqrt{lg}}\right) \leq \arccos \left( \frac {1} {p + 1} \right) $$

<br /><br />

$$\frac{um}{2M\sqrt{lg}} \leq \sin \left( \frac {\arccos \left( \frac {1} {p + 1} \right)} {2}\right)$$

<br /><br />


$$M \geq \frac {um} {2 \sqrt{lg} \sin \left( \frac {\arccos \left( \frac {1} {p + 1} \right)} {2}\right)}$$

Результат получился реалистичным - при достаточно большой массе, угол отклонения будет достаточно маленьким для требуемых приближенных равенств, потому нас интересует нижняя оценка этой массы, которую мы и получили.

Для практического рассчета, положим $l$=1 м, $m$ = 0.01 кг, $u$ = 15 м/с. Будем варьировать параметр $p$, и смотреть, как от него зависит минимальная требуемая масса:

```python
def min_mass(u, m, l, p):
    sn = np.sin(np.arccos(1 / (p + 1))/ 2)
    
    return u * m / (2 * np.sqrt(l * g) * sn)
```

```python
args = np.linspace(0.005, 0.1)

func = np.vectorize(lambda x: min_mass(15, 0.01, 1, x))
plt.figure(figsize=[7, 7])

plt.title("Зависимость минимальной массы от требуемой точности")
plt.xlabel("Максимальная относительная погрешность")
plt.ylabel("Минимальная требуемая масса, кг")

plt.plot(args, func(args));
plt.grid();
```

Как видно из графика, для приемлимой погрешности $p=0.05$, достаточно взять маятник массой 150 грамм. Маятника массой 500 грамм хватит для того, чтобы свести ее практически на нет, для любых достаточно простых установок, которые не являются особо точными сами по себе.
